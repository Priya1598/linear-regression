```python
df2=pd.DataFrame(boston['data'],columns=boston.feature_names)
df2
```

```python
boston.target
```

```python
df2['housing_value']=boston.target
```

```python
df2
```

```python
X= df2.drop(['housing_value'],axis=1)
y=df2['housing_value']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
xtrain,xtest,ytrain,ytest
```

```python
xtrain
```

```python
ytrain
```

```python
len(xtrain)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
list(zip(df2.columns,model.coef_))
```

```python
df2.columns
```

```python
predictors=df2.columns[:-1]
predictors
```

```python
coef=pd.Series(model.coef_,df2.columns[:-1]).sort_values()
coef
```

```python
coef.plot(kind='bar',title='model coefficents')
```

```python
plt.scatter(df2.RM,df2.housing_value)
```

```python
plt.scatter(df2.NOX,df2.housing_value)
```

```python
print('R2 score :{}'.format(model.score(xtest,ytest)))
```

# prediction

```python
plt.scatter(ytrain,model.predict(xtrain))
```
